### tallest and shortest players

```python
def convert_height_to_cm(height):
    parts = height.split("'")
    feet, inches = int(parts[0]), int(parts[1].replace('"', ''))
    total_inches = feet * 12 + inches
    height_in_cm = total_inches * 2.54
    return height_in_cm

# Add a new column 'Height_cm' to the DataFrame
data['Height_cm'] = data['Height'].apply(convert_height_to_cm)
```

```python
max_height_row= data[['Name', 'Age','OVA','POT','Club','weight_in_kgs','Height_cm']].loc[data['Height_cm'].idxmax()]
max_height_row
```

```python
min_height_row= data[['Name', 'Age','OVA','POT','Club','weight_in_kgs','Height_cm']].loc[data['Height_cm'].idxmin()]
min_height_row
```
